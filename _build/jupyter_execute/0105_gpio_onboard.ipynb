## Tlačítko

Nestlačené tlačítko vráti hodnotu 1, stlačené 0.

```Python
from pyb import Pin
p = Pin('PC13', Pin.IN)

# tlac hodnoty tlacitka 0/1
print(p.value())  

# definovanie prerusenia a callback funkcie
p.irq(lambda p:print(p.value()))
```
