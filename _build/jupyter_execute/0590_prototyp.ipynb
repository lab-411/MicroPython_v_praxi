# Prototypy obvodov

Prevažná časť elektronických obvodov použitých v tejto knihe je založená na dostupných moduloch určených pre platformu Arduino. Ak potrebujeme vytvoriť vlastné zapojenie, môžeme použiť zapojovaciu dosku (bread-board) s prepojovacími vodičmi alebo obvod postaviť na univerzálnej doske.
