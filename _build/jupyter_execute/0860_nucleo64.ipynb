# STM32 Nucleo-64 

Vývojový modul STM32 Nucleo-64 môže byť osadený rôznymi typmi procesorov triedy STM32 v 64-pinovom púzdre LQFP. Na module sú vyvedené konektory pre pripojenie periférií kompatibilných s *Arduino UNO* a konektory s vyvedenými pinmi mikrokontroléra označené ako *ST morpho extension header*. Zapojenie pinov a elektrické zapojenie modulu je uvedené v  [dokumentácii](./doc/nucleo_64.pdf).


Pri inicializácií periférií môžeme použiť označenie pinov podľa konektorov *ST morpho* alebo konektorov *Arduino*.

## Periférie na module Nucleo-64 

| Periféria         | Nucleo-64 | Arduino-UNO | Poznámka
| :---              | :--       | :--         | :--  
| Zelená LED        | PA5       |             |
| Modré tlačítko    | PC13      |             |


## Rozhranie I2C 

| Periféria         | Nucleo-64 | Arduino-UNO | Poznámka
| :---              | :--       | :--         | :--  
| I2C1 SCL          | PB8       |             |
| I2C1 SDA          | PB9       |             |
| I2C2 SCL          | PB10      |             |
| I2C2 SDA          | PB11      |             |
| I2C3 SCL          | PC0       |             |
| I2C3 SDA          | PC1       |             |


## Rozhranie SPI </font>

| Periféria         | Nucleo-64 | Arduino-UNO | Poznámka
| :---              | :--       | :--         | :--  
| SPI1 NSS          | PA4       | A2          | NC D10
| SPI1 SCK          | PB3       | D3          | NC D13
| SPI1 MISO         | PB4       | D5          | NC D12
| SPI1 MOSI         | PB5       | D4          | NC D11
| SPI2 NSS          | PB12      |             | Nepripojený
| SPI2 SCK          | PB13      |             |
| SPI2 MISO         | PB14      |             |
| SPI2 MOSI         | PB15      |             |


## Časovač </font>

| Periféria         | Nucleo-64 | Arduino-UNO | Poznámka
| :---              | :--       | :--         | :--  
| TIM2 CH3          | PB10      | D6          |
| TIM3 CH1          | PB4       | D5          |
| TIM3 CH2          | PC7       | D9          |



## Rozhranie USART </font>

| Periféria         | Nucleo-64 | Arduino-UNO | Poznámka
| :---              | :--       | :--         | :--  
| USART1 TX         | PA9       | D8          |
| USART1 RX         | PA10      | D2          |
| USART2 TX         | PA2       | D1          | NC, SB63
| USART2 RX         | PA3       | D0          | NC, SB62
| USART2 RTS        | PA1       | A1          |
| USART2 CTS        | PA0       | A0          |
| USART3 TX         | PC4       |             |  
| USART3 RX         | PC6       |             |  
| USART3 RTS        | PB14      |             |
| USART3 CTS        | PB13      |             |
| USART4 TX         | PA0       | A0          |
| USART4 RX         | PA1       | A1          |
| USART5 TX         | PC12      |             |
| USART5 RX         | PD2       |             |
| LPUART1 TX        | PA12      |             |
| LPUART1 RX        | PA11      |             |    

## CAN Bus </font>

| Periféria         | Nucleo-64 | Arduino-UNO | Poznámka
| :---              | :--       | :--         | :--  
| CAN1 TX           | PA12      |             |
| CAN1 RX           | PA11      |             |


## ADC </font>

| Periféria         | Nucleo-64 | Arduino-UNO | Poznámka
| :---              | :--       | :--         | :--  
| ADC IN1           | PC0       | A5          |
| ADC IN2           | PC1       | A4          |
| ADC IN3           | PC2       |             |
| ADC IN4           | PC3       |             |


Mapovanie pinov GPIO je uvedené v súbore [pins_L476.csv](./doc/pins_L476.csv). Mapovanie interných periférií je v súbore [stm32l476_af.csv](./doc/stm32l476_af.csv).
